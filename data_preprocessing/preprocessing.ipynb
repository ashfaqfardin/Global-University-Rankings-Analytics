```python
import pandas as pd
```

```python
the_world_ranking = pd.read_csv('../data/the_world_ranking.csv')
the_world_ranking.head()
```

```python
df = the_world_ranking.copy()


df[['University Name', 'Country']] = df['Name'].str.split(pat='\n', n=1, expand=True)
df['University Name'] = df['University Name'].str.strip()
df['Country'] = df['Country'].str.strip()


df['Overall Rank or Rank Band'] = df['Rank'].astype(str).str.replace(r'^=', '', regex=True).str.strip()

df = df.rename(columns={
    'Teaching': 'Teaching Score',
    'Research Environment': 'Research Environment Score',
    'Research Quality': 'Research Quality Score',
    'International Outlook': 'International Outlook Score',
    'Industry': 'Industry or Income Score'
})

cols = [
    'University Name',
    'Country',
    'Overall Rank or Rank Band',
    'Teaching Score',
    'Research Environment Score',
    'Research Quality Score',
    'International Outlook Score',
    'Industry or Income Score'
]
cleaned_rankings = df[cols]

cleaned_rankings.head()
```

```python
initial_count = len(cleaned_rankings)
cleaned_rankings.dropna(subset=cols, inplace=True)
cleaned_rankings.reset_index(drop=True, inplace=True)
print(f"Removed {initial_count - len(cleaned_rankings)} rows with null values")
cleaned_rankings.head()
```

```python
cleaned_rankings['Numeric Rank'] = cleaned_rankings['Overall Rank or Rank Band'].str.extract(r'(\d+)', expand=False).astype(int)
cleaned_rankings.head()
```
